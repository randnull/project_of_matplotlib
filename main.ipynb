Отреставрировать старинный график используя Matplotlib 

Для реставрирования я выбрал график из газеты "The Wall Street Journal" от 19-го октября 1987 года. Информация в этой газете было о резком падении индекса Доу-Джонса (DJIA).

![picture](photo.jpg)

Используя документацию Matplotlib, был реализован код, который записывает информацию о месте, куда на картинке был сделан клик.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mimg

img = mimg.imread('input.jpg')

fig, ax = plt.subplots()
ax.imshow(img)
data_list_x = []
data_list_y = []

def onclick(event):
    data_list_x.append(event.x)
    data_list_y.append(event.y)
    ax.scatter(event.xdata, event.ydata, color='red')
    plt.draw()


cid = fig.canvas.mpl_connect('button_press_event', onclick)

plt.show()
```
